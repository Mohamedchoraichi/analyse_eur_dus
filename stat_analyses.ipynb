```python
import numpy as np
import pandas as pd
```

```python
#import data
data = pd.read_csv("EURUSD.csv",names=['Date','Heure' , 'Open', 'High' , 'Low',  'Close' , 'Volume'])
```

```python
data.head(5)
```

```python
#Tabeau contenant tous les jours sans prise en copte la date 
date_en_jour=data['Date'].unique()
```

```python
date_en_jour
```

```python
data['prix']= (data['High']+ data['Low'])/2 #pour le prix on a pris la moyenne entre high and low
```

```python
data = data.reindex(columns = data.columns.tolist() + ['prix_moyen']) #rajouté une colonne pour stocker le prix moyen
```

```python
from datetime import date ,time, datetime
dict={}
dtableau_valeurs=[]
for jour in date_en_jour:
#on va itirer  sur les jours et calculer les valeurs  dont on  a besoin  prix moyen variance valeur max hauss
    
    data_de_ce_jour=data.loc[data['Date'] == jour]
   
    prix_moyen = data_de_ce_jour['prix'].mean()
    variance_= data_de_ce_jour['prix'].var()
    valeur_max = max(data_de_ce_jour['High'])
    hauss= max(data_de_ce_jour['High'])-min(data_de_ce_jour['Low'])  
    #pour la hauus on a prix la difference entre la valeur max attient et la valeur plus faible dans une  journnée
    
    
    Jour = datetime.strptime(jour, "%Y.%m.%d")
    lejour = date.weekday(Jour)
    lesJours = ['Lundi', 'Mardi', 'Mercredi', 'Jeudi', 'Vendredi', 'Samedi', 'Dimanche']
    
    
    dict[jour]= prix_moyen
    dtableau_valeurs.append([lesJours[lejour],variance_,hauss,valeur_max])
```

```python
#creer colonne moyen
moyenne = []

for i in data["Date"]:
    moyenne.append(dict[i])
data["prix_moyen"] = moyenne    
data
```

```python
#creer une dataframe conteant chaque jour avec sa varianve  sa hauss et sa valuer maxiaml
new_data =pd.DataFrame(dtableau_valeurs, columns =['jour','variance','hauss','valeur_max'])
```

```python
max(new_data['hauss'])
```

```python
jour_de_semaine=new_data['jour'].unique()
```

```python
tableau_final=[]
for jour in jour_de_semaine:
    data_de_ce_jour=new_data.loc[new_data['jour'] == jour]
    
    moyen_de_la_variance=data_de_ce_jour['variance'].mean()
    plus_grand_hauss= max(data_de_ce_jour['hauss'])
    valeur_max= max(data_de_ce_jour['valeur_max'])
    
    tableau_final.append([jour,moyen_de_la_variance,valeur_max,plus_grand_hauss])
```

```python
data_final=pd.DataFrame(tableau_final,columns=['jour','moyenne_variance','valeur_max','plus_grand_hauss'])
```
